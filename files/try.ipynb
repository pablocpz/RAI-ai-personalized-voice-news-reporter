```python
narrative = main(news_content)

from openai import OpenAI

client = OpenAI()

# Generate speech
response = client.audio.speech.create(
    model="tts-1",
    voice="nova",
    input=narrative
)
print("------------------")
print(type(response))

# Check response for audio content (URL or binary data)
if hasattr(response, 'content') and response.content:
    with open("narrative.wav", "wb") as audio_file:
        audio_file.write(response.content)
    print("Audio saved successfully")
elif hasattr(response, 'url'):
    # If the API provides a URL for audio, you can download the file from there
    audio_url = response.url
    # Downloading logic for URL
    import requests
    audio_data = requests.get(audio_url).content
    with open("narrative.wav", "wb") as file:
        file.write(audio_data)
    print("Audio downloaded from URL")
else:
    print("No valid audio content found in the response.")
```
